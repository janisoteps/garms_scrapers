```python
cat_dict_arr = re.findall('(?<=NAV_ENTRY_TYPE_CATEGORY).*?(?=\})', ajax_req.text)
```

```python
cat_dict_arr
```

```python
cat_name_arr = [re.search('(?<=label\"\:\").*?(?=\")', cat_dict).group(0) for cat_dict in cat_dict_arr]
```

```python
cat_name_arr
```

```python
cat_filter_arr = [re.search('(?<=categoryFilter\"\:\").*?(?=\")', cat_dict).group(0) for cat_dict in cat_dict_arr]
```

```python
cat_filter_arr
```

```python
cat_info_arr = [{'cat_name': cat_name, 'cat_filter': cat_filter_arr[cat_name_arr.index(cat_name)]} for cat_name in cat_name_arr]
```
