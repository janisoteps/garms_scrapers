```python
word_dict = {}
```

```python
for prod_dict in cat_list:
    words = prod_dict['prod_name'].split(' ')
    for word in words:
        lower_word = word.lower()
        if lower_word in word_dict:
            word_dict[lower_word] += 1
        else:
            word_dict[lower_word] = 1
```

```python
word_list = word_dict.items()
```
