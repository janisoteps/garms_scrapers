```python
homepage_req = requests.get('https://www.farfetch.com/uk/', headers=headers)
homepage_doc = lxml.html.fromstring(homepage_req.text)
```

```python
cat_url_els = homepage_doc.xpath('.//ul[contains(@class, "ff-primary-nav")]//a[contains(@class, "ff-nav-a")]')
```

```python
def get_sex(url_string):
    if 'women' in url_string:
        return 'women'
    elif 'men' in url_string:
        return 'men'
    else:
        return None
```

```python
cat_urls = []
for cat_url_el in cat_url_els:
    cat_url_match = cat_url_el.xpath('@href')[0]
    if len(cat_url_match.split('?')) > 1:
        cat_url = f'https://www.farfetch.com{cat_url_match.split("?")[0]}'
    else:
        cat_url = f'https://www.farfetch.com{cat_url_match}'
    cat_name = cat_url_el.xpath('text()')[0].strip()
    cat_sex = get_sex(cat_url)
    if len(cat_name) > 0 and cat_sex is not None:
        cat_urls.append({
            'url': cat_url,
            'name': cat_name,
            'sex': cat_sex
        })
```

```python
cat_req = requests.get('https://www.farfetch.com/uk/shopping/women/sale/all/items.aspx', headers=headers)
cat_doc = lxml.html.fromstring(cat_req.text)
```

```python
prod_tiles = cat_doc.xpath('.//li[@data-test="productCard"]')
```

```python
prod_list = []
for prod_tile in prod_tiles:
    prod_url = prod_tile.xpath('.//a[@itemprop="itemListElement"]/@href')[0]
    prod_name = prod_tile.xpath('.//p[@data-test="productDescription"]/text()')[0]
    price_match = prod_tile.xpath('.//span[@data-test="price"]/text()')[0]
    initial_price = prod_tile.xpath('.//span[@data-test="initialPrice"]/text()')[0]
    sale = False
    # price = None
    saleprice = None
    if initial_price is not None:
        sale = True
        price = initial_price
        saleprice = float(price_match.replace('¬£', ''))
    else:
        price = price_match
    brand = prod_tile.xpath('.//h3[@data-test="productDesignerName"]/text()')[0]

    prod_list.append({
        'name': prod_name.title(),
        'prod_url': 'https://www.farfetch.com' + prod_url,
        'price': float(price.replace('¬£', '')),
        'sale': sale,
        'saleprice': saleprice,
        'brand': brand
    })
```
