```python
def get_sex(url_string):
    women_check = len(url_string.split('women')) == 2
    if women_check == True:
        return 'women'
    else:
        men_check = len(url_string.split('men')) == 2
        if men_check == True:
            return 'men'
        else:
            return None
    
```
