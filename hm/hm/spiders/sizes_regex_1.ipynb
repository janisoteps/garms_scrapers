```python
import re
input_file = open('test_2.txt', 'r')
input_string = input_file.read()
```

```python
m = re.search('(?<=productArticleDetails\ =\ )[\s\S]*(?=\;)', input_string)
```

```python
prod_id = '0579302004'
```

```python
prod_match = re.search(f'(?<=\'{prod_id}\'\:\ )[\s\S]*(?=\,\n\        \')', m.group(0))
```

```python
prod_match.group(0)
```

```python
sizes_match = re.search('(?<=\'sizes\'\:\[)[\s\S]*?(?=\]\,)', prod_match.group(0))
```

```python
sizes_match.group(0)
```

```python
size_name_match = re.findall('(?<=\"name\"\:\ \")[\S]*?(?=\")', sizes_match.group(0))
```

```python
size_name_match
```

```python
sizes_stock = [{
                'stock': 'In stock',
                'size': size
            } for size in size_name_match]
```
