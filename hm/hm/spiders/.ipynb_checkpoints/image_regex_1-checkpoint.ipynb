```python
import re
input_file = open('test_2.txt', 'r')
input_string = input_file.read()
```

```python
m = re.search('(?<=productArticleDetails\ =\ )[\s\S]*(?=\;)', input_string)
```

```python
m.group(0)
```

```python
prod_id = '0579302004'
```

```python
m2 = re.search(f'(?<=\'{prod_id}\'\:\ )[\s\S]*(?=\,\n\        \')', m.group(0))
```

```python
m2.group(0)
```

```python
images_string_match = re.search('(?<=\'images\'\:\[)[\s\S]*?(?=\]\,\n)', m2.group(0))
```

```python
# m2
```

```python
images_string_match.group(0)
```

```python
# m2.group(0)
```

```python
m3 = re.findall('(?<=fullscreen\'\:\ isDesktop\ \?\ \').*?(?=\'\ \:)', images_string_match.group(0))
```

```python
m3
```

```python
imageUrls = [f'http:{img_slug}' for img_slug in m3]
```
