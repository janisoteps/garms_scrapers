```python
home_req = requests.get('https://www.collusion.com/collection', headers=headers)
```

```python
home_doc = lxml.html.fromstring(home_req.text)
```

```python
cat_matches = home_doc.xpath('.//ul[@class="collection"]/li/ul/li/a')
```

```python
cat_dicts = []
for cat_match in cat_matches:
    cat_url = f'https://www.collusion.com{cat_match.xpath(".//@href")[0]}'
    cat_name = cat_match.xpath('.//text()')[0]
    cat_dicts.append({
        'cat_url': cat_url,
        'cat_name': cat_name
    })
```
